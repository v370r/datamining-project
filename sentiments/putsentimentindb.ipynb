```python
con = sqlite3.connect("processdDB.db")

df = pd.read_sql("SELECT * FROM processdDB", con)

# reports = []
# for i,group in df.groupby("name"):
#     for j,report in group.groupby("time"):
#         reports.append(report[report["type"]=="10-K"])
#         #a = report[report["type"]=="10-K"].copy()
#         #b = report[report["type"] == "10-Q"].copy()
#         #reports.append(a)
#         #reports.append(b)
```

```python
tags = ["RF","URFC","Comments","Cybersecurity","Properties","MDAFC","QQDMR","FSSD"]
sans = SentimentAnalyzer()
index_sentiment = {}

for target in tags:
    tdf = df[df["tag"] == target]
    print(f"Current tag {target}")
    for i,row in tqdm(tdf.iterrows(),total=len(tdf)):
        sans.add_text(row["value"])
        index_sentiment[i] = sans.get_sentiment()
```
